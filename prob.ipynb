```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data for plotting
p = np.arange(0.01, 0.99, 0.001)
H = -p * np.log2(p) - (1 - p) * np.log2(1 - p)

fig, ax = plt.subplots()
ax.plot(p, H)

ax.set(xlabel='p', ylabel='Entropia',
       title='Entropia de evento com dois poss√≠veis valores')
ax.grid()

# fig.savefig("test.png")
plt.show()
```
