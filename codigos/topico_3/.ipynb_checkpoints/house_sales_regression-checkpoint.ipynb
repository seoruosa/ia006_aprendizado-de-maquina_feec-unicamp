```python
import pandas as pd
import numpy as np
filepath = '~/code/ia006_aprendizado-de-maquina_feec-unicamp/codigos/topico_3/kc_house_data.csv'

house_data = pd.read_csv(filepath)
# print(house_data.describe())

# print(house_data.columns)
# print(house_data[['price','sqft_lot']].describe())

y = house_data[['price']].values
x = house_data[['sqft_lot']].values

phi = np.concatenate((np.ones(x.shape),x),1)
w_best = np.linalg.solve(phi.T @ phi, phi.T @ y)
print(w_best)
```

```python
# Make data.
X = np.arange(500000, 600000, 500)
Y = np.arange(0, 1, 0.01)
X, Y = np.meshgrid(X, Y)

shape_ini = X.shape
X = X.reshape(shape_ini[0]*shape_ini[1],1)
Y = Y.reshape(shape_ini[0]*shape_ini[1],1)

W = np.concatenate((X,Y),1)
phi = np.concatenate((np.ones(x.shape),x),1)

def Jerror(w, x, y, phi):
    return ( y.T @ y + y.T @ phi @ w + w.T @ phi.T @ y + w.T @ phi.T @ phi @ w);

e = np.zeros((W.shape[0],1))
for i in range(e.shape[0]):
    e[[i]] = Jerror(W[i],x,y,phi)

X = X.reshape(shape_ini)
Y = Y.reshape(shape_ini)
Z = e.reshape(shape_ini)
# print(e)
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
%matplotlib inline

fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
# ax.set_zlim(-1.01, 1.01)
# ax.zaxis.set_major_locator(LinearLocator(10))
# ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
